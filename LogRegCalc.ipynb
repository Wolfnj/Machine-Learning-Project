```r
data =  subset(data, select = -c(15))
# This column had too many NAs
```

```r
data =  subset(data, select = -c(1))
head(data)
```

```r
#Splitting data into training and test data sets
data_obs = nrow(data)
data_idx = sample(data_obs, size = trunc(0.80 * data_obs))
##Truncate 1/2*iris_obs to get half the total observations in a whole number

data_trn = data[data_idx, ]
data_tst = data[-data_idx, ]
```

```r
head(data_trn)
```

```r
#install.packages("nnet")
library(nnet)
model_multi = multinom(Type ~ ., data = data_trn)
summary(model_multi)$coefficients
```

```r
#summary(model_multi)
```

```r
head(predict(model_multi, newdata = data_trn))

head(predict(model_multi, newdata = data_trn, type = "prob"))
```

```r
#We store the predictions made on the train and test sets.
data_log_trn_pred = predict(model_multi, newdata = data_trn, type="class")
data_log_tst_pred = predict(model_multi, newdata = data_tst, type="class")
```

```r
head(data_log_trn_pred,n=200)
```

```r
head(data_log_tst_pred,n=200)
```

```r
calc_class_err = function(actual, predicted) {
mean(actual != predicted)
}
```

```r
#Training misclassification error
calc_class_err(predicted = data_log_trn_pred, actual = data_trn$Type)
#Test misclassification error
calc_class_err(predicted = data_log_tst_pred, actual = data_tst$Type)
```

```r
#Confusion matrix
table(predicted = data_log_tst_pred, actual = data_tst$Type)
```
